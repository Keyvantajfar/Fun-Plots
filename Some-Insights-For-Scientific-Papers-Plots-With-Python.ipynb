*   ### Great! Now its time for some beautiful plots

```python
# We will go for a bidirectional (side by side) horizontal barchart.
import matplotlib as mpl
import seaborn as sns
#sns.set()
#%matplotlib notebook
#plt.style.use('classic')

class_list = ['test1','test2','test3','test4','test5','test6','test7']
average_length = [1, 2, 3]
num_entries = [1, 2, 3]
data = pd.DataFrame(data=zip(class_list,average_length,num_entries),columns=['Class','Lens','Nums'])
data.set_index('Class', inplace=True)

font_color = '#525252'
hfont = {'fontname':'Calibri'}
#facecolor = '#eaeaf2'
color_red = '#fd625e'
color_blue = '#01b8aa'
index = data.index
column0 = data['Lens']
column1 = data['Nums']
title0 = "Title 1"
title1 = 'Title 2'

# fig, axes = plt.subplots(figsize=(10,5), facecolor=facecolor, ncols=2,sharey=True)
fig, axes = plt.subplots(figsize=(10,5), ncols=2,sharey=True)
fig.tight_layout()

axes[0].barh(index, column0, align='center', color=color_red, zorder=10)
axes[0].set_title(title0, fontsize=18, pad=15, color=color_red, **hfont)
axes[1].barh(index, column1, align='center', color=color_blue, zorder=10)
axes[1].set_title(title1, fontsize=18, pad=15, color=color_blue, **hfont)


# If you have positive numbers and want to invert the x-axis of the left plot
axes[0].invert_xaxis() 


# To show data from highest to lowest
plt.gca().invert_yaxis()

axes[0].set(yticks=data.index, yticklabels=data.index)
axes[0].yaxis.tick_left()
axes[0].tick_params(axis='y', colors='white') # tick color

# axes[1].set_xticklabels(['0','20', '40', '60', '80', '100', '120'])


for label in (axes[0].get_xticklabels() + axes[0].get_yticklabels()):
    label.set(fontsize=13, color=font_color, **hfont)
for label in (axes[1].get_xticklabels() + axes[1].get_yticklabels()):
    label.set(fontsize=13, color=font_color, **hfont)
    
plt.subplots_adjust(wspace=0, top=0.85, bottom=0.1, left=0.18, right=0.95)

# To save your plot in png format you can uncomment the line below
# plt.savefig('Plot3.png', dpi='figure', bbox_inches='tight')
plt.show()
```

*   ### And finally we go for a 3D plot!

```python
from mpl_toolkits.mplot3d import Axes3D

# Create the dataframe from the given data
data = {
    'Method': methods,
    'DataSet': ['ml-1M', 'ml-1M', 'ml-1M', 'ml-1M', 'ml-1M', 'ml-1M', 'ml-1M', 'ml-1M'],
    'Metric of sensitive attribute Example @5': [0.7182, 1.8100, 1.9552, 2.7499, 1.4862, 1.7988, 14.8250, 3.6335],
    'Metric of sensitive attribute Example @10': [3.2507, 4.3906, 1.8937, 4.2980, 1.8724, 1.1146, 16.3889, 4.3940],
    'Metric of sensitive attribute Example @15': [3.4805, 4.2363, 3.3609, 5.0553, 1.9794, 2.7732, 14.7978, 5.0231]
}
df = pd.DataFrame(data)

# Prepare the data for plotting
methods = df['Method']
topK = ['@5', '@10', '@15']
x_pos = range(len(topK))
y_pos = range(len(methods))

X, Y = np.meshgrid(x_pos, y_pos)  # Use np.meshgrid from NumPy

Z = [
    [df.loc[i, f'Metric of sensitive attribute Example {topK[j]}'] for j in range(len(topK))] 
    for i in range(len(methods))
]

# Create the figure and axis
fig = plt.figure(figsize=(18, 12))
ax = fig.add_subplot(111, projection='3d')

######################################################
# Plot the 3D bars Colourized
# You can use any of the provided colour pallets.
#
# plot_color_gradients('Perceptually Uniform Sequential',
#                      ['viridis', 'plasma', 'inferno', 'magma', 'cividis'])
# plot_color_gradients('Sequential',
#                      ['Greys', 'Purples', 'Blues', 'Greens', 'Oranges', 'Reds',
#                       'YlOrBr', 'YlOrRd', 'OrRd', 'PuRd', 'RdPu', 'BuPu',
#                       'GnBu', 'PuBu', 'YlGnBu', 'PuBuGn', 'BuGn', 'YlGn'])
# cmap = plt.get_cmap('hsv')
cmap = plt.get_cmap('viridis')
colors = ['red', 'blue', 'green', 'orange']
dx = 0.4
dy = 0.8
for i in range(len(methods)):
    dz = np.array(Z[i])
    ax.bar3d(X[i], Y[i], 0, dx, dy, dz.ravel(), color=cmap(i / float(len(methods))), alpha=0.7)

# Set the labels and title
ax.set_xlabel('TopK', labelpad=15)
ax.set_ylabel('Models', labelpad=20)
ax.set_zlabel('Metric', labelpad=10)
ax.set_title('Metric of sensitive attribute Example for ml')

# Set the ticks and tick labels
ax.set_xticks(x_pos)
ax.set_xticklabels(topK)
ax.set_yticks(y_pos)
ax.set_yticklabels(methods)

# Rotate the ticks and labels
ax.set_xticklabels(topK, rotation=45)
# ax.set_yticklabels(methods, rotation=45)
ax.set_yticklabels(methods)

# Set view angle
ax.view_init(elev=30, azim=-50)

# Adjust the layout
fig.tight_layout()

# To save your plot in png format you can uncomment the line below
# plt.savefig('MovieLensRelativeAge-3D.png', dpi='figure', bbox_inches='tight')

# Show the plot
plt.show()
```
